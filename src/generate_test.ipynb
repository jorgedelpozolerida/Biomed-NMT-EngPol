```python
import pandas as pd
```

# Merging

```python
file_path = "../data/khresmoi-summary-test-set-2.0/khresmoi-summary-dev.en"

with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

eng_dev = pd.DataFrame({'eng': lines, 'split' : 'dev'})
eng_dev
```

```python
file_path = "../data/khresmoi-summary-test-set-2.0/khresmoi-summary-dev.pl"

with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

pl_dev = pd.DataFrame({'pol': lines, 'split' : 'dev'})
pl_dev
```

```python
file_path = "../data/khresmoi-summary-test-set-2.0/khresmoi-summary-test.en"

with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

eng_test = pd.DataFrame({'eng': lines, 'split' : 'test'})

starting_value = 500
new_index = range(starting_value, starting_value + len(eng_test))
eng_test = eng_test.set_index(pd.Index(new_index))

eng_test 
```

```python
file_path = "../data/khresmoi-summary-test-set-2.0/khresmoi-summary-test.pl"

with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

pol_test = pd.DataFrame({'pol': lines, 'split' : 'test'})

starting_value = 500
new_index = range(starting_value, starting_value + len(pol_test))
pol_test = pol_test.set_index(pd.Index(new_index))

pol_test 
```

```python
frames = [pl_dev, pol_test]
pol = pd.concat(frames)
pol
```

```python
frames = [eng_dev, eng_test]
eng = pd.concat(frames)
eng
```

```python
duplicated_indices = eng.index[eng.index.duplicated()]
print("Duplicated Indices:", duplicated_indices)
pol = pol.drop(columns='split')
```

```python
df = pd.concat([pol, eng], axis=1, join='inner')
data = df
data['id'] = df.index
df.insert(0, 'id', df.pop('id'))
data
```

# Preprocessing

```python
filter_out = data.loc[data["pol"] == data["eng"]]
filter_out_ids = filter_out[["id"]]
filter_out_ids
```

```python
data = data[~data.id.isin(filter_out_ids.id.to_list())]
data
```

```python
filter_out_polish = data[data['pol'].apply(lambda x: len(x) < 15)]
filter_out_english = data[data['eng'].apply(lambda x: len(x) < 15)]
filter_out_ids_polish = filter_out_polish[["id"]]
filter_out_ids_english = filter_out_english[["id"]]
```

```python
filter_out_ids_polish
```

```python
filter_out_ids_english
```

```python
data = data[~data.id.isin(filter_out_ids_polish.id.to_list())]
data = data[~data.id.isin(filter_out_ids_english.id.to_list())]
data
```

```python
# Regex pattern for Cyrillic characters
pattern = r'[Яа-яЁё]'

# Filter out rows with Russian sentences
data = data[~data['pol'].str.contains(pattern)]
# Filter out rows with Russian sentences
data = data[~data['eng'].str.contains(pattern)]
```
