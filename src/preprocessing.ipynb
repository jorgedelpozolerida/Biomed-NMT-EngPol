```python
import pandas as pd
```

```python
data = pd.read_csv("../data/medical_corpus_clean.tsv")
```

```python
data
```

# Remove all rows where Polish text is identical to English text


```python
filter_out = data.loc[data["pol"] == data["eng"]]
filter_out_ids = filter_out[["id"]]
filter_out_ids
```

```python
data = data[~data.id.isin(filter_out_ids.id.to_list())]
data
```

# Remove all rows where Polish OR English text has less than 15 characters

```python
filter_out_polish = data[data['pol'].apply(lambda x: len(x) < 15)]
filter_out_english = data[data['eng'].apply(lambda x: len(x) < 15)]
filter_out_ids_polish = filter_out_polish[["id"]]
filter_out_ids_english = filter_out_english[["id"]]
```

```python
filter_out_ids_polish
```

```python
filter_out_ids_english
```

```python
data = data[~data.id.isin(filter_out_ids_polish.id.to_list())]
data = data[~data.id.isin(filter_out_ids_english.id.to_list())]
data
```

# Try to filter out russian sentences based on regex characters unique to cyryllic

```python
# Regex pattern for Cyrillic characters
pattern = r'[Яа-яЁё]'

# Filter out rows with Russian sentences
data = data[~data['pol'].str.contains(pattern)]
# Filter out rows with Russian sentences
data = data[~data['eng'].str.contains(pattern)]
```

```python
data
```

# Try to filter out french sentences based on regex characters unique to french language

```python
# Regex pattern for Cyrillic characters
pattern = r'[ÉéÀàÈèÙùÂâÊêÎîÔôÛûÇçËëÏïÜü]'

# Filter out rows with Russian sentences
data = data[~data['pol'].str.contains(pattern)]
# Filter out rows with Russian sentences
data = data[~data['eng'].str.contains(pattern)]
```
