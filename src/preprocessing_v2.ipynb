```python
import pandas as pd
```

```python
data = pd.read_csv("../data/medical_corpus_clean.tsv")
```

```python
data
```

# Remove all rows where Polish text is identical to English text


```python
filter_out = data.loc[data["pol"] == data["eng"]]
filter_out_ids = filter_out[["id"]]
filter_out_ids
```

```python
data = data[~data.id.isin(filter_out_ids.id.to_list())]
data
```

# Remove all rows where Polish OR English text has less than 15 characters

```python
filter_out_both = data[data.apply(lambda row: len(''.join(c for c in str(row['pol']) if c.isalpha())) < 15 or len(''.join(c for c in str(row['eng']) if c.isalpha())) < 15, axis=1)]

# If you only want the 'id' column from the filtered DataFrame
filter_out_ids_both = filter_out_both[["id"]]
```

```python
filter_out_ids_both
```

```python
data = data[~data.id.isin(filter_out_ids_both.id.to_list())]
data
```

# Try to filter out russian sentences based on regex characters unique to cyryllic

```python
# Regex pattern for Cyrillic characters
pattern = r'[Яа-яЁё]'

# Filter out rows with Russian sentences
data = data[~data['pol'].str.contains(pattern)]
# Filter out rows with Russian sentences
data = data[~data['eng'].str.contains(pattern)]
```

```python
data
```

# Try to filter out french/greek/bulgarian/romanian sentences based on regex characters unique to french language

```python
# Regex pattern for French characters
pattern = r'[ÉéÀàÈèÙùÂâÊêÎîÔôÛûÇçËëÏïÜü]'

# Filter out rows with French sentences
data = data[~data['pol'].str.contains(pattern)]
# Filter out rows with French sentences
data = data[~data['eng'].str.contains(pattern)]


# Define a pattern for Greek, Bulgarian, and Romanian alphabets
greek_bulgarian_romanian_pattern = r'[Α-Ωα-ωА-Яа-яĂăÂâÎîȘșȚț]'

# Filter out rows with Greek, Bulgarian, and Romanian sentences
data = data[~data['pol'].str.contains(greek_bulgarian_romanian_pattern)]
# Filter out rows with Greek, Bulgarian, and Romanian sentences
data = data[~data['eng'].str.contains(greek_bulgarian_romanian_pattern)]
```
