```python
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
file_path = "../LUMI/logs/train/5088823.output"
```

```python
# Cell 3: Read and parse data
try:
    with open(file_path, 'r') as file:
        # Read the content of the file
        file_content = file.readlines()

    time_list = []
    loss_list = []
    eval_loss_list = []

    for line in file_content:
        try:
            entry = eval(line)  # Evaluate the string as a Python expression
            if 'epoch' in entry:
                # Assuming 'epoch' values are already in seconds
                epoch_time = entry['epoch']
                time_list.append(epoch_time)
                loss_list.append(entry.get('loss', 0))
                eval_loss_list.append(entry.get('eval_loss', 0))
        except (SyntaxError, ValueError):
            # Ignore lines that are not valid Python dictionaries
            pass

except FileNotFoundError:
    print(f"The file '{file_path}' does not exist.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
result = [(value, eval_loss_list[i]) for i, value in enumerate(time_list) if eval_loss_list[i] != 0]
epochs, evals = zip(*result)

print("Epochs:", epochs)
print("Values:", evals)
```

```python
#loss_list = loss_list[:7]
#time_list = time_list[:7]
#eval_loss_list = eval_loss_list[:7]
```

```python
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

plt.plot(time_list, loss_list, label='Training Loss')
plt.plot(epochs, evals, label='Eval Loss')

plt.xlabel('epoch')
plt.ylabel('Loss')
plt.title('100K')
plt.legend()

# Rotate x-axis text
plt.xticks(rotation=45, ha='right')  # Adjust the rotation angle and alignment as needed

plt.tight_layout()  # Ensures that labels are not cut off
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have 'time_list', 'loss_list', 'epochs', and 'evals' defined somewhere in your code

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot the training loss on the left y-axis
ax1.plot(time_list, loss_list, label='Training Loss', color='blue')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Training Loss', color='blue')
ax1.tick_params('y', colors='blue')

# Create a twin Axes sharing the xaxis
ax2 = ax1.twinx()

# Plot the eval loss on the right y-axis
ax2.plot(epochs, evals, label='Eval Loss', color='red')
ax2.set_ylabel('Eval Loss', color='red')
ax2.tick_params('y', colors='red')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('100K')

# Rotate x-axis text
plt.xticks(rotation=45, ha='right')  # Adjust the rotation angle and alignment as needed

plt.tight_layout()  # Ensures that labels are not cut off
plt.show()
```
